```python
from ttkwidgets.autocomplete import AutocompleteEntry
from tkinter import *
greetings = ['Hello How Are You', 'where is the lab', 'I like students',
'Best Wishes','Thanks and Regards', 'United States of America',
"Prince Shri Venkateswara Padmavathy Engineering College"]
```

```python
greetings
```

```python
ws = Tk()
ws.title('Python-AUTO-COMPLETE-example')
ws.geometry('400x300')
ws.config(bg='#f25252')
frame = Frame(ws, bg='#f25252')
frame.pack(expand=True)
Label(
frame,
bg='#f25252',
font = ('Times',21),
text='MY MESSAGES'
).pack()
entry = AutocompleteEntry(
frame,
width=30,
font=('Times', 36),
completevalues=greetings
)
entry.pack()
```

```python
ws.mainloop()
```
